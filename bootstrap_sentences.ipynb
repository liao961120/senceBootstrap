```python
## Get stopwords
with open('stop_words.txt') as f:
    stopwords = f.readlines()
stopwords = [x.strip() for x in stopwords]
stopwords.append('<')
stopwords.append('>')
stopwords = set(stopwords)
'其他' in stopwords
```

## Bootstrapping Sentences

```python
for sense in cwn['V'].keys():
    
    bs_sentences = []
    for sentence in cwn['V'][sense]['examples']:
        keys = seed2keys(sentence, stopwords)
        bs_sentences.append(bootstrap(target_word=keys[0], keys=keys, corp_text=corp_text, l_window=30, r_window=15))
    
    bs_sentences = list(flatten(bs_sentences))
    cwn['V'][sense]['bs_sentences'] = bs_sentences
```
