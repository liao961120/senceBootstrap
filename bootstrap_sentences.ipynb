## 語料庫 (平衡語料庫 + zhWiki)

```python
with open('asbc5_corp.txt') as f:
    corp_text = f.readlines()
corp_text = ''.join(corp_text)
corp_text = corp_text.replace(u'\u3000',u'')
corp_text = corp_text.replace('\n', '')
corp_text0 = corp_text
```

```python
with open('zhwiki_nopunc.txt') as f:
    corp_text1 = f.read()
```

## CWN 資料

```python
import json

with open('sense_anno/cwn_annot_Jessica_0321.json', 'r') as f:
    cwn = json.load(f)
```

### Stop Words: words with $log(\frac{content~words~freq.}{function~words~freq.}) < 0$

```python
## Get stopwords
with open('stop_words.txt') as f:
    stopwords = f.readlines()
stopwords = [x.strip() for x in stopwords]
stopwords.append('<')
stopwords.append('>')
stopwords = set(stopwords)
```

## Bootstrapping Sentences

### 平衡語料庫

```python
from seed_sentences_processing import seed2keys, seed2keys2, flatten
from bootstrap_sentences_from_corp import bootstrap

for sense in cwn['V'].keys():
    bs_sentences = []
    
    for sentence in cwn['V'][sense]['examples']:
        keys = seed2keys(sentence, stopwords)
        print(keys)
        bs_sentences.append(bootstrap(target_word=keys[0], keys=keys[1:], corp_text=corp_text0, l_window=25, r_window=25))
    
    bs_sentences = list(flatten(bs_sentences))
    cwn['V'][sense]['bs_sentences'] = bs_sentences
```

```python
for sense in cwn['V'].keys():
    print(cwn['V'][sense]['bs_sentences'])
```

### Inspection of Corpus

```python
from random import shuffle, seed
from bootstrap_sentences_from_corp import find_all_target_word_window

freq_汽車 = len(find_all_target_word_window('汽車', corp_text0, l_window=15, r_window=15))
freq_車 = len(find_all_target_word_window('車', corp_text0, l_window=15, r_window=15))
freq_汽 = len(find_all_target_word_window('汽', corp_text0, l_window=15, r_window=15))

print('汽車：', freq_汽車)
print('車：', freq_車)
print('汽：', freq_汽)
```

#### Use Character Based Search

`seed2keys2()` is used instead of `seed2keys()`.  
`seed2keys2()` returns a string of 'reduced' sentence (not a list of keywords) from a CWN example sentence. The length of the string is deternined by windows size.

```python
import json
with open('sense_anno/cwn_annot_Jessica_0321.json', 'r') as f:
    cwn = json.load(f)
    
for sense in cwn['V'].keys():
    bs_sentences = []
    
    for sentence in cwn['V'][sense]['examples']:
        keys = seed2keys2(sentence, stopwords, l_window=4, r_window=4)
        bs_sentences.append(bootstrap(target_word=keys[0], keys=keys[1], corp_text=corp_text0, l_window=25, r_window=25))
    
    bs_sentences = list(flatten(bs_sentences))
    cwn['V'][sense]['bs_sentences'] = bs_sentences

for sense in cwn['V'].keys():
    print(cwn['V'][sense]['bs_sentences'])
```
